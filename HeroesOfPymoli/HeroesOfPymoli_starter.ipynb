```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python

player_count = len(purchase_data["SN"].unique())
total_players = pd.DataFrame({'Total Players': [player_count], })
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = purchase_data["Item ID"].value_counts()
unique_items_counts=(len(unique_items))
average_price = purchase_data["Price"].mean()
number_of_purchases = purchase_data["Purchase ID"].count()
total_revenue = purchase_data["Price"].sum()

summary_table = pd.DataFrame({"Number of Unique Items": unique_items_counts,
                              "Average Price": average_price,
                              "Number of Purchases":number_of_purchases,
                              "Total Revenue":[total_revenue]})
summary_table["Average Price"] = summary_table["Average Price"].apply('{:.2f}'.format)
summary_table["Total Revenue"] = summary_table["Total Revenue"].apply('${:,.2f}'.format)
summary_table
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_table = purchase_data[["SN", "Gender"]]
gender_table = gender_table.drop_duplicates()
gender_table.drop(["SN"], axis=1)

gender_count = gender_table["Gender"].value_counts()
percentage_of_players =(gender_count/player_count)*100

summary_table_gender = pd.DataFrame({"Total Count":gender_count,"Percentage of Players":percentage_of_players})
summary_table_gender =summary_table_gender.round(2)
summary_table_gender
```


## Purchasing Analysis (Gender)

* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
demo_table = purchase_data[["SN", "Age"]]
demo_table = demo_table.drop_duplicates()

#Create bins in which to place values 
bins = [0, 9, 14, 19, 24, 29, 34, 39,100]

# Create labels for these bins
group_labels = ['<10' ,'10-14', '15-19' , '20-24' ,'25-29', '30-34','35-39', '40+']

# Place the data series into a new column inside of the DataFrame
demo_table["Age Group"] = pd.cut(demo_table["Age"], bins, labels=group_labels)

age_group = demo_table.groupby("Age Group")

# Find how many rows fall into each bin
tot_unique_count =age_group["SN"].count()


# Get the average of each column within the GroupBy object
percentage_of_count =(tot_unique_count/player_count)*100
percentage_of_count

summary_table_demo = pd.DataFrame({"Total Count":tot_unique_count,"Percentage of Players":percentage_of_count})
summary_table_demo = summary_table_demo.round(2)
summary_table_demo
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_table = purchase_data

#Create bins in which to place values 
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]

# Create labels for these bins
group_labels = ['<10', '10-14' ,'15-19', '20-24' ,'25-29' ,'30-34','35-39', '40+']

# Place the data series into a new column inside of the DataFrame
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=group_labels)

purchase_c = purchase_table.groupby("Age Group")
tot_count =purchase_c["SN"].count()
avg_price =purchase_c["Price"].mean()
sum_price =purchase_c["Price"].sum()
avg_total_purchase_per_person =  (sum_price/tot_count)
summary_table_purchase = pd.DataFrame(
                                    {"Purchase Count":tot_count,
                                     "Average Purchase Price":avg_price,
                                     "Total Purchase Value":sum_price})
summary_table_purchase = summary_table_purchase.round(2)
summary_table_purchase["Avg Total Purchase per Person"] = summary_table_purchase["Total Purchase Value"] / tot_unique_count

summary_table_purchase_first = summary_table_purchase[0:1].copy()
summary_table_purchase_rest = summary_table_purchase[1:].copy()
final_table = summary_table_purchase_rest.append(summary_table_purchase_first)
final_table["Average Purchase Price"] = final_table["Average Purchase Price"].apply('${:,.2f}'.format)
final_table["Total Purchase Value"] = final_table["Total Purchase Value"].apply('${:,.2f}'.format)
final_table["Avg Total Purchase per Person"] = final_table["Avg Total Purchase per Person"].apply('${:,.2f}'.format)
final_table
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_data_new = purchase_data[["SN", "Price"]]
spender_table = purchase_data_new.groupby(["SN"])
summary_table_spender = spender_table.agg(['count', 'mean', 'sum'])
top_spender = summary_table_spender.Price.sort_values(['sum'], ascending=False)

new_summary_table= top_spender.rename(columns={
              "count": "Purchase Count", "mean": "Average Purchase Price", "sum":"Total Purchase Value"})
new_summary_table =new_summary_table.round(2)
new_summary_table.head()
new_summary_table["Average Purchase Price"] = new_summary_table["Average Purchase Price"].apply('${:.2f}'.format)
new_summary_table["Total Purchase Value"] = new_summary_table["Total Purchase Value"].apply('${:.2f}'.format)

new_summary_table.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_item_group = purchase_data.groupby(["Item ID", "Item Name"])
popular_item_table = popular_item_group .agg(['count', 'last', 'sum'])
popular_item_table = popular_item_table.Price.sort_values(['count'], ascending=False)

popular_item_table_summary_table = popular_item_table.rename(columns={"count": "Purchase Count", "last": "Item Price", "sum":"Total Purchase Value"})
popular_item_table_summary_table.head()
```
